```python
def plus(price):
    return price+price

class MixFunctions:
    def __init__(self, price, discount):
        self.price: int = price
        self.discount: int = discount
        self.double = plus(self.price)
        
    def get_discounted_price(self) -> int:
        discounted_rate = self.discount/100
        discounted_price = self.price * (1 - discounted_rate)
        return int(discounted_price)
    
    def return_double_price(self) -> int:
        return self.double
    
```

```python
product_price = 11000
discount_pct = 15

pay = MixFunctions(product_price, discount_pct).get_discounted_price()
print(f"{product_price}원에서 {discount_pct}% 할인 적용되어 {pay} 원")
```
